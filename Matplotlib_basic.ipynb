### Style

-Marker:
- . : point marker
- , : pixel marker
- o : circle marker
- v : triangle_down marker
- ^ : triangle_up marker
- < : triangle_left marker
- > : triangle_right marker
- 1 : tri_down marker
- 2 : tri_up marker
- 3 : tri_left marker
- 4 : tri_right marker
- s : square marker
- p : pentagon marker
- * : star marker
- h : hexagon1 marker
- H : hexagon2 marker
- + : plus marker
- x : x marker
- D : diamond marker
- d : thin_diamond marker

- 라인 스타일
- - : solid line style
- -- : dashed line style
- -. : dash-dot line style
- : : dotted line style

- 커스텀 스타일
- c : 선 색상
- lw : 선 굵기
- ls : 선 스타일
- marker : 마커 종류
- ms : 마커 크기
- mec : 마커 선 색깔
- mew : 마커 선 굵기
- mfc : 마커 내부 색깔

```python
data=np.random.randint(10,size=10)
plt.plot(data,'r*:')#red, star point, dot line
plt.show()
```

```python
data=np.random.randint(10,size=10)
plt.plot(data,c='b',lw=2,ls='-',marker='o',ms='20',mec='r',mew='5',mfc='y')
#c='b' color is blue
#lw=2 선굵기 2
#ls='-' 선 스타일은 -
#marker='o' 마커 스타일은 o
#ms=20 마커 크기는 20
#mec=r 마커 선 색깔은 red
#mew=5 마커 선 굵기는 5
#mfc=y 마커 내부 색깔은 yellow
#plt.plot에 대한 help을 통해 알아가면 된다!
plt.show()
```

### Bar plot

```python
y=np.random.randint(10,size=5)
x=range(5)
plt.bar(x,y)
plt.show()
```

```python
y=np.random.randint(5,10,size=5)
x=range(5)
error=np.random.randint(1,5,size=5)
print(y,error)
plt.bar(x,y,yerr=error,align='edge',alpha=0.5)
#yerr는 말 그대로 error값을 받는다.일종의 confidence level
# align은 xtick위의 bar가 해당 point의 center에 있는지, edge에 있는지를 나타낸다.
#alpha는 bar 색의 농도수준을 의미
plt.show()
```

```python
y=np.random.randint(5,10,size=5)
x=range(5)
error=np.random.randint(1,5,size=5)
print(y,error)
plt.barh(x,y,xerr=error,align='center',alpha=0.6)
plt.show()
#barh은 말그대로 horizontal bar chart를 의미한다~!!
```

## scatter plot

```python
#uniform (균등분포)
x1=np.random.rand(500)
y1=np.random.rand(500)
plt.subplot(331)
plt.scatter(x1,y1)

#gaussian(정규분포)
x2=np.random.randn(500)
y2=np.random.randn(500)
plt.subplot(332)
plt.scatter(x2,y2)

#randint(uniform)
x3=np.random.randint(1,100,size=500)
y3=np.random.randint(1,100,size=500)
plt.subplot(333)
plt.scatter(x3,y3)

plt.show()
```

```python
data1=np.random.randint(1,10,size=20)
data2=np.random.randint(1,10,size=20)
colors=np.random.rand(20)
area=np.pi*np.arange(10,50,2)**2 #area of circle
plt.scatter(data1,data2,s=area,c=colors,alpha=0.5)
#s=area// s는 marker의 size를 의미한다. 해당 소스코드에서는 랜덤 반지름의 원의 크기
#c=colors//c는 marker의 color를 의미한다. 해당 소스코드에서는 0~20까지의 랜덤 숫자
plt.show()
```

### histogram

```python
#50개의 구간으로 잘라 해당되는 데이터의 수를 카운트한 후에 y축의 값을 normalize
x1=np.random.rand(1000)
plt.subplot(211)
plt.hist(x1,bins=50)

x2=np.random.randn(1000)#히스토그램이 normal distribution의 형태를 띈다.
plt.subplot(212)
plt.hist(x2,bins=200)

plt.show()
#bins=50//여기서 x데이터가 1000개 반환되었다. bins=50이기때문에 histogram 1000/50=20개
#의 Bar가 나오게 된다. 즉, 더 촘촘해진다.
```
