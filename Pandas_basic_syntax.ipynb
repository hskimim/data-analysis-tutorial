## Merge
- make DataFrame with ID, Name data
- make DataFrame with ID, Money data

```python
#make data functions
# make data which includes name and age

import random,string

def get_name():
    names=['Adam', 'Alan', 'Alex', 'Alvin', 'Andrew', 'Anthony', 'Arnold', 'Baldy']
    return random.choice(names)

def get_age(start=20,end=40):
    return random.randint(start,end)
        
def make_data(rows=10):
    datas=[]
    for _ in range(rows):
        data= {"Age": get_age(), "Name": get_name()}
        datas.append(data)
    return datas
```

```python
# make user_df
user_df = pd.DataFrame(columns=["UserID","Name","Age"])

for idx in range(1,9):
    name = get_name()

    #make name data with no overlapped
    while name in list(user_df["Name"]):
        name = get_name()
    
    # data insert
    data = {"Name":name, "UserID":idx, "Age":get_age()}
    user_df.loc[len(user_df)] = data
    
user_df    
```

```python
money_df=pd.DataFrame(columns=["ID","Money"])

for idx in range(15):
    money=random.randint(1,20)*1000
    data={"Money":money,"ID": random.randint(1,10)}
    money_df.loc[len(money_df)]=data
    
money_df
```

```python
#match moeny_df's ID column and user_df's UserID column and merge
# datas in UserID and ID are same, but columns' names are different so two cols are 
# inserted in dataframe.
# default of how is inner 
money_df.merge(user_df,left_on="ID",right_on="UserID")
```

```python
money_df.merge(user_df,left_on="ID",right_on="UserID",how="outer")
```

```python
#In user_df, change UserID name into ID and merge
# we dont have to right down 'left_on' and 'right_on'
user_df.rename(columns={"UserID":"ID"},inplace=True)
user_df
```

```python
result_df=pd.merge(money_df,user_df)
result_df
```

```python
#groupby Name and sum the Money and reset index
money_list=result_df.groupby("Name").agg('sum').reset_index()
#sort the Money data in descending and reset index
money_list=money_list.sort_values(by=['Money'],ascending=False).reset_index(drop=True)
money_list
```

```python
#if you use 'outer' in how, then Money data which is NaN returns 0 with fillna(value=0)
result=pd.merge(user_df,money_list,how='outer').fillna(value=0)
result=result.sort_values(by=["Money"],ascending=False).reset_index(drop=True)
result
```

```python
#chage data type
#converting datatype of Money column which is float datatype.
result["Money"]=result["Money"].astype('int')
result
```

- Name: A,B,C,D,E
- Point: random 60~100
- Let's make DataFrame!

```python
df=pd.DataFrame(columns=["Name","Point"])
df["Name"]=["A","B","C","D","E"]
df["Point"]=np.random.randint(60,100,size=5)
df
```

```python
def chg(point):
    if point >=90:
        return 'A'
    elif point >=80:
        return 'B'
    elif point >=70:
        return 'C'
    elif point >=60:
        return 'D'
    else:
        return 'F'
df['Result']=df["Point"].apply(chg)
df
```
