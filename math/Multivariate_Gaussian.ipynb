## 다변수 가우시안 정규 분포(Multivariate Gaussian normal dist)
- 다변수 가우시안 정규 분포(MVN)는 복수의 확률 변수를 모형화하는데 가장 많이 쓰이는 분포입니다.
- D차원 multivariate pdf는 Mu와 sigma matrix를 모수로 가지며 다음과 같은 수식으로 정의됩니다.
- N(x;Mu,Sigma) = 1/((2*pi)^D/2 * det(Sigma)^1/2 ) *exp(-1/2*(x-Mu).T * inv(Sigma) * (x-Mu))
- 위의 등식은 이전에 다뤘던 단변수 가우시안 정규 분포에서 scalar value가 vector로 확장된 것입니다.
    - x : R_D rv vector
    - Mu : R_D mean vector
    - Sigma : R_DxD Covariance matrix
    - Sigma inverse : R_DxD Inversed Covariance matrix (precision matrix,concentration matrix)

### example of MVN
- mu = np.array([2],[3])
- sigma = np.array([1,0],[0,1])

```python
mu=[2,3]
cov = [[1,0],[0,1]]
rv=sp.stats.multivariate_normal(mu,cov)
xx= np.linspace(0,4,120)
yy=np.linspace(1,5,150)
XX,YY = np.meshgrid(xx,yy)
plt.grid(False)
plt.contourf(XX,YY,rv.pdf(np.dstack([XX,YY])))
plt.axis('equal')
plt.show()
```
