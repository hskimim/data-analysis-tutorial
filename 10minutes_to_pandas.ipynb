## Grouping
- groupby 함수에 내장되어 있는 arguments들입니다! 읽어보시고 예제에서 확인해보세요
    - Splitting = 어떠한 기준으로 데이터가 그룹 별로 나뉘게 됩니다!
    - Applying = 함수들이 각각의 그룹으로 독립적으로 적용됩니다!
    - Combining = 자료형에 결과값으로 들어갑니다

```python
df=pd.DataFrame({
    'A': ['foo', 'bar', 'foo', 'bar','foo','bar','foo','foo'],
    'B':['one','one','two','three','two','two','one','three'],
    'C':np.random.randn(8),
    'D':np.random.randn(8)
                })
```

```python
df
```

grouping을 column A 기준으로 해보겠습니다(foo and bar)
그리고 그 기준으로 나눈 values들을 더해서 DataFrame으로 나타내겠습니다.

```python
df.groupby('A').sum()
```

이번에는 grouping을 A(foo and bar),B(oneltwo,three) 두 개의 criteria를 적용하고 그 값들을 더해서 DataFrame으로 나타내겠습니다.

```python
df.groupby(['A','B']).sum()
```

```python
tuples=list(zip(*[['bar','bar','baz','baz',
                 'foo','foo','qux','qux'],
                ['one','two','one','two',
                'one','two','one','two']]))
tuples
```

Convert list of tuples to MultiIndex
튜플들의 리스트를 멀티인덱스로 변환시켜주는 함수 MultiIndex.from_tuples를 사용한다

```python
df=pd.DataFrame(np.random.rand(8,2),index=index,columns=['A','B'])
df
```

```python
df2=df[:4]
df2
```

- stack() 메소드는 데이터프레임의 컬럼들의 레벨을 '압축'시켜준다.
- stack()과 unstack() 메소드는 서로 inverse한 관계로써,예시를 보면 알 수 있다.
- 컬럼 레벨을 데이터 프레임의 로우 레벨에 귀속시키므로써 계층적 인덱스로 만들어준다.
- 이 레벨들은 자동배열이 됩니다.(automatically get sorted.)

```python
stacked=df2.stack()
```

```python
stacked
```

```python
stacked.unstack()#default level= -1
```

## Pivot-Tables

```python
df = pd.DataFrame({'A' : ['one', 'one', 'two', 'three'] * 3,
                    'B' : ['A', 'B', 'C'] * 4,
                    'C' : ['foo', 'foo', 'foo', 'bar', 'bar', 'bar'] * 2,
                    'D' : np.random.randn(12),
                    'E' : np.random.randn(12)})
df
```

데이터 프레임을 이용해서 피벗 데이블을 손쉽게 만들 수 있습니다!

```python
pd.pivot_table(df,values='D', index=['A','B'], columns=['C'])
```

위를 보게 되면 values에 'D', index에 'A','B' 컬럼값에 'C'가 들어가 있는데, 데이터 프레임을 보게 되면, D에는 random 값들이, A,B에는 one two three//A,B,C 가 있고 마지막 C에는 bar, foo 가 있습니다. 이를 기준으로 피벗 데이블이 구성되는데, 여기서 주의할 점은 인덱스 값으로 쓰인 것이 아니라 하나의 value가 아니라 리스트로써 리스트의 뒤에 쓰일 수록 레벨이 깊어지는 것을 알 수 있습니다.

## Time Series
- 판다스는 빈도 변환(frequency conversion) resampling에 있어서 간단하고 강력하고 효율적인 기능을 갖추고 있습니다.

```python
rng=pd.date_range('1/1/2012', periods=100,freq='S')
rng
# 고정된 데이터 시간빈도 인덱스를 반환한다. 빈도 디폴트 값은 하루(a day)이다.
```

```python
ts=pd.Series(np.random.randint(0,500,len(rng)), index=rng)
ts.tail()
```

```python
ts.resample('5Min').sum()
```

시간당 데이터(secondly data)를 5분 단위 데이터로(5-minutley) 변환시켜줍니다.
그리고 5분 단위 데이터의 random값들을 다 더해줍니다! groupby의 aggfunc느낌!!

```python
rng=pd.date_range('3/6/2012 00:00', periods=5, freq='D')
rng
```

```python
ts=pd.Series(np.random.randn(len(rng)), rng)
ts
```

```python
ts_utc=ts.tz_localize('UTC')
ts_utc
#UTC란 영어로는 Coordinated Universal Time으로 협정 세계시를 나타냅니다!
#Localize tz-naive TimeSeries to target time zone.
# tz_localize method로 tart time zone으로 위치화시킵니다!!
```

```python
ts_utc.tz_convert('US/Eastern')
# Convert tz-aware axis to target time zone.
# time zone(tz)를 변환시킵니다!
```

```python
rnd=pd.date_range ('1/1/2012', periods=5, freq='M' )
ts=pd.Series(np.random.randn(len(rng)), index=rng)
ts
```

#### 여기서 잠깐 다시 짚고 넘어가자!!
- Series: 
    - ds=pd.Series(data=None, index=None, dtype=None, name= None)

```python
ps=ts.to_period()
ps
# Convert Series from DatetimeIndex to PeriodIndex with desired
# frequency (inferred from index if not passed)
#위에서 선언한 ts는 Series형 자료형에 rnd라는 data range method를 가지고 있는 rnd를 
# 데이터를 value로 쓰고 있습니다. 여기서 ps가 to_period method를 통해서 이를 
# datatimeindex로 변환시키고 있습니다.
```

```python
ps.to_timestamp()
# Cast to datetimeindex of timestamps, at *beginning* of period
```

period와 timestamp 간의 변환은 산술적 연산을 편하게 해주는 기능이 있습니다. 다음 예세에서 우리는 매년이 11월 9시에 끝나는 4분기 빈도수의(Q-NOV) 시간대를 매 달이 끝나는 4분기 시간대의 오전 9시로 변환하겠습니다.asfreq('M','e') + 1 asfreq('D','s') + 9
무슨 말인지 잘 모르겠죠?? 저도 쓰면서 잘 모르겠습니다.. 일단 예제를 보시죠!!

```python
prng=pd.period_range('1990Q1','2000Q4',freq='Q-NOV')
prng
```

```python
ts=pd.Series(np.random.randn(len(prng)), prng)
ts.tail()
```

```python
ts.index=(prng.asfreq('M','e')+1).asfreq('H','s') + 9
ts.tail()
#ts의 인덱스를 바꿔줍니다!!
#어떻게?? : 빈도수: month , end +1 , hour, start +1
```
