```python
import numpy as np
import pandas as pd
import scipy as sp
import matplotlib as mpl
import matplotlib.pylab as plt
import seaborn as sns
```

```python
train_16 = pd.read_csv('train_2016_v2.csv',parse_dates=['transactiondate'])
```

```python
prop_16 = pd.read_csv('properties_2016.csv')
```

```python
train_16.head()
```

```python
train_df = pd.merge(train_16,prop_16,on='parcelid', how ='left')
```

```python
sns.countplot(train_df.calculatedbathnbr,data=train_df)
plt.show()
```

```python
print(train_df.calculatedbathnbr.describe()),
print('length of calculatedbathnbr:{}'.format(len(train_df.calculatedbathnbr)))
```

```python
print(train_df.threequarterbathnbr.describe()),
print('length of threequarterbathnbr:{}'.format(len(train_df.threequarterbathnbr)))
```
